```python
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
url = "https://www.fastcampus.co.kr/"
```

```python
driver = webdriver.Chrome('C:\chromedriver\chromedriver.exe')
# driver = webdriver.PhantomJS('C:\phantomjs-2.1.1\bin\phantomjs.exe')
```

* get() : 특정 사이트로 이동할 수 있도록 드라이버에게 요청

```python
driver.get(url)
```

* XPath는 XML의 특정 요소를 지정할 때 사용
* CSS 선택자는 CSS로 요소를 디자인할 때 사용하는 표기 방법
* CSS 선택자가 보기 편하고 간편하다
